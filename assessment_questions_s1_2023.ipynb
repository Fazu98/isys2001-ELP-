```
from newspaper import Article
from typing import List, Dict

def extract_article_info(url_list: List[str]) -> List[Dict[str, str]]:
    articles_info = []
    for url in url_list:
        try:
            # Initialize Article object
            article = Article(url)
            # Download the article
            article.download()
            # Parse the article
            article.parse()
            # Add article information to the list
            articles_info.append({"URL": url, "Title": article.title, "Text": article.text})
        except Exception as e:
            print(f"An error occurred with URL {url}. Error: {e}")
    return articles_info
```



This script takes a list of URLs as input. For each URL, it creates an Article object from the newspaper3k library, then tries to download and parse the article. The title and text of the article are then extracted and added to a dictionary along with the URL. This dictionary is added to a list, articles_info, which is returned by the function.

newspaper3k is a Python library used for web scraping purposes to extract articles and perform natural language processing on text. It simplifies the task of extracting useful data from a webpage. In this script, it's used to download and parse the article at each URL and then to extract the title and full text of the article.

If a URL doesn't point to a valid article, or if there are network errors, the download() or parse() methods will raise an exception. This script catches any such exceptions, prints an error message, and then continues with the next URL.
